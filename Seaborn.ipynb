```python
fmri = sns.load_dataset('fmri')
fmri.head(10)
```

```python
# Lineplot with 95% Confidence Interval
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line')
```

```python
# Lineplot with Standard Deviation as Confidence Interval
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci='sd')
```

```python
# Lineplot without Confidence Interval
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', estimator=None)
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', style='region', hue='event')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', style='region', hue='event', markers=True, dashes = False)
```

```python
# Conditions
```

```python
sns.relplot(x='timepoint', y='signal', hue='region', units='subject', estimator=None, kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal', hue='region', units='subject', estimator=None, kind='line', data=fmri.query("event=='stim'"))
```

```python
# unit='subject ==> For each subject there is a different line.'
```
