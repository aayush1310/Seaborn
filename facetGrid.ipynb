```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# FacetGrid Allows us to create lots of Mini Graph
```

```python
sns.FacetGrid(tips, col='smoker')
```

```python
# Two Graphs 
# One for Non- Smoker and One for Smoker
```

```python
g = sns.FacetGrid(data=tips, col='smoker', row='time')
g = g.map(sns.distplot, 'total_bill')
```

```python
h=sns.FacetGrid(data=tips, col='smoker', row='time')
h=h.map_dataframe(sns.scatterplot, 'total_bill', 'tip', size='size').set_axis_labels('Total Bill', 'Tip')
```

```python
g=sns.FacetGrid(tips, col='size', col_wrap=3)
g=(g.map(sns.distplot,'tip', color='c', kde=False).set_titles("{col_name} diners"))
```

```python
i = sns.FacetGrid(data=tips, col='smoker', row='day')
i = (i.map_dataframe(sns.scatterplot,'total_bill','tip')).set_axis_labels('Total Bill', 'Tip')
```
